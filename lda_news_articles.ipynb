```python
def lemmatize(text):
    return WordNetLemmatizer().lemmatize(text, pos='v')
```

```python
lemmatize("गोयल")
```

```python
def rem_user(t):
    t = re.sub('@[^\s]+','',t)
    return re.sub('@[^\s]+','',t)
```

```python
def spellcheck(text):
    return  re.sub(r'[^a-z]', '', text)
```

```python
def remove_url(text):
    result = re.sub(r"http\S+", "", text)
    return result
```

```python
def preprocess(text):
    text = remove_url(text)
    text = rem_user(text)
    result = []
    for token in gensim.utils.simple_preprocess(text, True):
        if token not in gensim.parsing.preprocessing.STOPWORDS:
            lem = lemmatize(token)
            if len(lem)>3:
                result.append(lem)
            
    return result
```

```python
gensim.utils.simple_preprocess("वित्त मंत्री AŚ पीयूष गोयल ने कहा कि सरकार ")
```

```python
processed_docs = document
processed_docs['text'] = processed_docs['text'].map(preprocess)
```

```python
dictionary = gensim.corpora.Dictionary(processed_docs['text'])
count = 0
for k, v in dictionary.iteritems():
    print(k, v)
    
```

```python
dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
```

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs['text']]
bow_corpus[410]
```

```python
bow_doc_43 = bow_corpus[23]
for i in range(len(bow_doc_43)):
    print("Word {} (\"{}\") appears {} time.".format(bow_doc_43[i][0], 
                                               dictionary[bow_doc_43[i][0]], 
bow_doc_43[i][1]))
```

```python
from gensim import corpora, models


tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]

from pprint import pprint

for doc in corpus_tfidf:
    pprint(doc)
    break
```

```python
lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=10, id2word=dictionary, passes=2, workers=2)
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
import csv
with open("results_bow_news.csv", 'w') as writeFile:
    writer = csv.writer(writeFile)
    writer.writerows([['Topic','Values']])
    total = []
    for idx,topic in lda_model.print_topics(-1):
    #     print (topic)
        topiclist = []
        topiclist.append(str(idx))
        for x in topic.split("+"):
            topiclist.append(str(x.split("*")[1].replace("\"", "").replace(" ", "")))
        print("Topic: ",topiclist)
        total.append(topiclist)
    writer.writerows(list(total))
writeFile.close()
```

```python
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, num_topics=10, id2word=dictionary, passes=2, workers=4)
for idx, topic in lda_model_tfidf.print_topics(-1):
    print('\nTopic: {}\t \nWord: {}'.format(idx, topic))
```
