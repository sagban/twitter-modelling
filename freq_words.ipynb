```python
import preprocess_data
import re
from collections import Counter

def concat(docs):
    string = ""
    for doc in docs:
        for w in doc:
            string= string+w + " "

            
    return string

def freq(string):
    return Counter(re.findall(r"([A-Za-z]+'[A-Za-z]|[A-Za-z]+)", string))

```
